```python
# Step 1: 安裝 Whisper 和必要套件
!pip install git+https://github.com/openai/whisper.git
!apt update && apt install ffmpeg
!pip install pydub

# Step 2: 上傳音檔
from google.colab import files
uploaded = files.upload()  # 選擇你的音檔（例：xxx.mp3）

# Step 3: 將音檔切成 15 分鐘一段
from pydub import AudioSegment

filename = list(uploaded.keys())[0]  # 取得上傳檔名
audio = AudioSegment.from_file(filename)

chunk_length_ms = 15*60*1000  # 15 分鐘 = 15*60*1000 毫秒
chunks = [audio[i:i+chunk_length_ms] for i in range(0, len(audio), chunk_length_ms)]

chunk_files = []
for idx, chunk in enumerate(chunks):
    chunk_name = f"chunk_{idx}.mp3"
    chunk.export(chunk_name, format="mp3")
    chunk_files.append(chunk_name)

print(f"已切成 {len(chunk_files)} 段，每段約 15 分鐘")

# Step 4: 逐段使用 Whisper 轉錄
import whisper

model = whisper.load_model("small")  # 可以改 base / medium / large
all_text = ""

for idx, chunk_name in enumerate(chunk_files):
    print(f"正在轉錄第 {idx+1} 段: {chunk_name} ...")
    result = model.transcribe(chunk_name, language="zh")  # 強制中文
    all_text += result["text"] + "\n\n"

# Step 5: 將完整文字存成檔案
with open("transcription.txt", "w", encoding="utf-8") as f:
    f.write(all_text)

print("✅ 轉錄完成！文字已存成 transcription.txt")

# Step 6: 下載文字檔
files.download("transcription.txt")
```
